## 4.2 Escolha do Teste Estatístico

### 4.2.1 Teste t de Student

O teste t de Student é um teste estatístico paramétrico usado para determinar se há uma diferença significativa entre as médias de dois grupos. É amplamente utilizado quando os dados são aproximadamente normalmente distribuído e as variâncias dos grupos são homogêneas.

* Hipótese Nula (H0): Não há diferença significativa entre as médias dos grupos.
* Hipótese Alternativa (H1): Há uma diferença significativa entre as médias dos grupos.

Suposições que precisam ser atendidas:
* Normalidade
* Independência das Amostras
* Homogeneidade das Variâncias (ou Homocedasticidade)

#### 4.2.1.1 Validando as Suposições do Teste Paramétrico

## Teste de Normalidade

```python
# Teste de Shapiro-Wilk para normalidade
shapiro_A = stats.shapiro(grupo_A['conversao'])
shapiro_B = stats.shapiro(grupo_B['conversao'])
```

```python
print(shapiro_A)
print(shapiro_B)
```

* Hipótese Nula (H0): Os dados são normalmente distribuídos.
* Hipótese Alternativa (H1): Os dados não são normalmente distribuídos.

Se o valor-p for menor que α, rejeitamos a hipótese nula. Caso contrário, não rejeitamos a hipótese nula.

* Para o Grupo A, o valor p é significativamente menor que 0.05. Isso significa que rejeitamos a hipotese nula de que os dados são normalmente distribuídos. Portanto, os dados do Grupo A não são normalmente distribuídos.
* Para o Grupo B, o valor p é significativamente menor que 0.05. Isso significa que rejeitamos a hipotese nula de que os dados são normalmente distribuídos. Portanto, os dados do Grupo A não são normalmente distribuídos.

Ações a Tomar

Como os dados de ambos os grupos não são normalmente distribuídos, a suposição de normalidade para o teste t de Student não é atendida. Neste caso, pode considerar os seguintes passos:
* transformação dos dados para aproximá-los de uma distribuição normal (por exemplo, transformações logarítmicas)
* usar um teste não paramétrico

## Teste de Homogeneidade das variâncias

```python
levene_test = stats.levene(grupo_A['conversao'], grupo_B['conversao'])
levene_test
```

* Hipótese Nula (H0): As variâncias dos dois grupos são iguais (homogeneidade das variâncias)
* Hipótese Alternativa (H1): As variâncias dos dois grupos são diferentes (heterogeneidade das variâncias)

Se o valor-p for menor que α, rejeitamos a hipótese nula. Caso contrário, não rejeitamos a hipótese nula.

o valor do p é maior que 0.05, por isso rejeitamos a hipótese nula.

## RESUMO
